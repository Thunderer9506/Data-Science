## Data Cleaning

- Many columns have Null values we need to fix it
- we need to drop __director__ column cause it contain many null values
- for other column they are needed to be deleted
- date_added need to be converted into datetime data type

```python
data.info()
```

```python
data.drop('director',axis=1,inplace=True)
```

```python
data.info()
```

Dropping Nan values

```python
data.dropna(axis=0,inplace=True,subset=['date_added'])
```

```python
data['date_added'] = pd.to_datetime(data['date_added'])
```

## Exploratory Data Analysis

### How Movies and TV Shows are there and Visualize it

```python
typeOfContent = ['Movie','TV Show']
values = list(data['type'].value_counts())

plt.figure(figsize=(8, 6)) # Set figure size

# Create the bar chart with styling
plt.bar(typeOfContent, values,
        color=['skyblue', 'lightgreen'], # Different colors for each bar
        width=0.7, # Adjust bar width
        edgecolor='black') # Add black borders


plt.xlabel('Type', fontsize=12)
plt.ylabel('Values', fontsize=12)
plt.title('Movie vs TV Show', fontsize=14, fontweight='bold')

# Add value labels on top of bars
for i, v in enumerate(values):
    plt.text(i, v, str(v), ha='center', va='bottom', fontsize=10)


plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```

### Count the no. of movies made in each country

Note: some rows contain many countries, pick only first one

```python
# Get the first country for each row
first_countries = data['country'].str.split(',', expand=True)[0]

# Count the number of movies per country
country_counts = first_countries.value_counts()

# Plot the top 20 countries for better readability
country_counts.head(20).plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.title('Number of Movies Made in Each Country (Top 20)')
plt.xticks(rotation=45, ha='right')

for i,v in enumerate(country_counts):
    if i <= 19:
        plt.text(i, v, str(v), ha='center', va='bottom', fontsize=7)
    else:
        break

plt.tight_layout()
plt.show()
```

### Show Time series data with date_added column

```python
moviesAdded = data['date_added'].value_counts().sort_index()

plt.figure(figsize=(12,7))
plt.title('Time Series of Date Movie Added')
plt.xlabel('Date')
plt.ylabel('Count')
plt.xlim((min(moviesAdded.index),max(moviesAdded.index)))
plt.ylim((min(moviesAdded),max(moviesAdded)+10))

plt.plot(moviesAdded.index,moviesAdded)
plt.show()
```

### Show how many movies were produced in each year gap like 1940-1960,1960-1980,etc.

```python
yearsData = data['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,7))
plt.title('No. of movies Release in each year gap',fontsize = 20)

plt.ylim((0,max(yearsData)+100))
plt.xlim((min(yearsData.index)-5,max(yearsData.index)+5))

plt.ylabel('Count',fontsize = 16)
plt.xlabel('Release Year',fontsize = 16)

plt.plot(yearsData.index,yearsData,marker='o')
plt.show()
```

### Show how many movies fall under each rating category

```python
ratingCount = data['rating'].value_counts()[:-3]

plt.figure(figsize=(10,5))
plt.title('No. of movie in each rating',fontsize = 16)

plt.bar(ratingCount.index,ratingCount)
plt.xticks(rotation=45,ha='right')

plt.ylabel('Count')
plt.xlabel('Movie Rating')

for i,v in enumerate(ratingCount):
    plt.text(i,v+3,str(v),ha='center',fontsize = 9)
plt.show()
```

### Split each category in listed_in column put in a dictionary and count no. of movies fall into each category and present it in graph

```python
category = {}

def categoryFunction(x):
    for i in x.split(', '):
        if i not in category:
            category[i] = 1
        else:
            category[i] += 1
    return category

data['listed_in'].apply(categoryFunction)
categoryData = pd.DataFrame([category],index=['Values']).transpose()
categoryData.sort_values(by='Values',ascending=False,inplace=True)
plt.figure(figsize=(13,7))
plt.title('No. of Movies/Show fall in each category',ha='right',fontsize = 16)

plt.barh(categoryData.index[0:20],categoryData['Values'][0:20])
plt.xlim((0,categoryData['Values'].max()+100))
plt.ylabel('Category')
plt.xlabel('Movies and Show')

plt.yticks(va='center')

plt.tight_layout()
plt.show()
```
